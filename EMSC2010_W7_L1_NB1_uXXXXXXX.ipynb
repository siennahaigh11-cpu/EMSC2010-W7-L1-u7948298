### Help a lost penguin find its way home

* During a winter field trip to Antartica you find a lost penguin.
* It's dark and there is a snow storm!
* You cannot see enough to identify what type of penguin it is.
* But you can measure its height, which is 71 cm
* Should you point in the penguin in the direction of the Emperor or Adélie colony?

What you know:
Heights of the Emperor penguin population are normally distributed:
* Mean = 115 cm
* Standard deviation = 5 cm

Heights of the Adélie penguin population are normally distributed:
* Mean = 60 cm
* Standard deviation = 8 cm

You also have Bayes theorem to help!

$p(A|B) = \frac{p(B|A)⋅p(A)}{p(B)}$,

where:

* $p(A)$ is the prior probability that the penguin is species $A$.
* $p(B|A)$ is the likelihood that a species $A$ penguin has a height of 71 cm.
* $p(B)$ is the probability of observing a 71 cm tall penguin (we won't worry about this, I'll explain why later).
* $p(A|B)$ is the posterior probability that the penguin is species $A$ given that its height is $B$ (this is what we are interested in).
